# New Keynesian Model with Endogenous Capital

We mostly follow Gali et al. (JMCB 2004) to model endogenous capital accumulation in a standard New Keynesian model. 

#### 1.1. The Household Problem

The household problem of this economy at time $t$ is given by: 

$$
\max_{ \{C_{t}, N_{t}, B_{t+1}, I_t, K_{t+1} \}_{t=0}^{\infty} } \mathbb E_0  \sum_{t=0}^{\infty} \beta^{t} U(C_{t}, N_{t})
$$

subject to the following budget constraint $\forall t$: 
$$
P_t (C_t + I_t) + \frac{1}{R_t} B_{t+1} = W_t N_t + R_{t}^k K_t + B_t + D_t
$$

and the law of motion for capital: 
$$
K_{t+1} = (1-\delta) K_t + \phi \left(\frac{I_t}{K_t} \right) K_t
$$

where $W_t$ is nominal wage, $R_t^k$ is the gross nominal return of renting capital, $B_t$ is the amount of riskless one-period nominal bonds carried over from time $t-1$ and paying one unit of the numeraire at time $t$, $R_t$ is the gross nominal return of bonds purchased at time $t$, $D_t$ are dividends from firm profits, and $\phi \left(\frac{I_t}{K_t} \right) K_t$ is the capital adjustment cost, where $\phi' \gt 0 $, $\phi'' \le 0$, $\phi'(\delta) = 1$, and $\phi (\delta) = \delta$. 

The Lagrangean of the household problem is given by: 
\begin{align*}
\mathcal{L} &= \mathbb E_0  \sum_{t=0}^{\infty} \beta^{t} \left[ U(C_{t}, N_{t}) 
                                                                + \lambda_{t} \bigg\{ W_t N_t + R_{t}^k K_t + B_t + D_t - P_t (C_t + I_t) - \frac{1}{R_t} B_{t+1} \bigg\} 
                                                                + \gamma_t \bigg\{ (1-\delta) K_t + \phi \left(\frac{I_t}{K_t} \right) K_t - K_{t+1} \bigg\} \right] \\
     &\equiv \mathbb E_0  \sum_{t=0}^{\infty} \beta^{t} \left[ u(c_{t}) - v(n_t) 
                                                                + \lambda_{t} \bigg\{ W_t N_t + R_{t}^k K_t + B_t + D_t - P_t (C_t + I_t) - \frac{1}{R_t} B_{t+1} \bigg\} 
                                                                + \gamma_t \bigg\{ (1-\delta) K_t + \phi \left(\frac{I_t}{K_t} \right) K_t - K_{t+1} \bigg\} \right] \\
     &\equiv \mathbb E_0  \sum_{t=0}^{\infty} \beta^{t} \left[ \frac{ C_{t}^{1-\sigma}-1}{1-\sigma} - \frac{N_{t}^{1+\psi}}{1+\psi} 
                                                                + \lambda_{t} \bigg\{ W_t N_t + R_{t}^K K_t + B_t + D_t - P_t (C_t + I_t) - \frac{1}{R_t} B_{t+1} \bigg\} 
                                                                + \gamma_t \bigg\{ (1-\delta) K_t + \phi \left(\frac{I_t}{K_t} \right) K_t - K_{t+1} \bigg\} \right] \\
\end{align*}

#### 1.2. Optimality Conditions for the Household Problem 

The FOCs are: 
\begin{alignat*}{2}
\text{[$C_{t}$] :} \quad & \frac{\partial u(C_{t})}{\partial C_{t}} = \lambda_{t} && \Longrightarrow \frac{C_{t}^{-\sigma}}{P_t} = \lambda_{t} \\
\text{[$N_{t}$] :} \quad & \frac{\partial v(N_{t})}{\partial N_{t}}  = \lambda_{t} W_t && \Longrightarrow N_{t}^\psi = \lambda_{t} W_t \\
\text{[$B_{t}$] :} \quad & \frac{\lambda_t}{R_t} = \beta  \mathbb E_t \lambda_{t+1} && \Longrightarrow \lambda_t = \beta R_t \mathbb E_t \lambda_{t+1} \\
\text{[$I_{t}$] :} \quad & \gamma_t \phi_{t}' = \lambda_t P_t \\
\text{[$K_{t+1}$] :} \quad & \gamma_t = \beta \mathbb E_t \lambda_{t+1} R_{t+1}^K + \beta \mathbb E_t \gamma_{t+1} \left[ 1 - \delta + \phi_{t+1} - \frac{I_{t+1}}{K_{t+1}} \phi_{t+1}' \right] 
\end{alignat*}

where $\phi_{t+1} \equiv \phi_{t+1} (\frac{I_{t+1}}{K_{t+1}} )$ and $\phi_{t+1}' \equiv \phi_{t+1}' (\frac{I_{t+1}}{K_{t+1}}) $

The CSCs are: 
\begin{align*}
& \lambda_{t} \bigg\{ W_t N_t + R_{t}^K K_t + B_t + D_t - P_t (C_t + I_t) - \frac{1}{R_t} B_{t+1} \bigg\} = 0, \\
& \lambda_{t} \geq 0, \\
& W_t N_t + R_{t}^K K_t + B_t + D_t - P_t (C_t + I_t) - \frac{1}{R_t} B_{t+1} \geq 0 \\
& \gamma_t \bigg\{ (1-\delta) K_t + \phi \left(\frac{I_t}{K_t} \right) K_t - K_{t+1} \bigg\} = 0 , \\
& \gamma_{t} \geq 0, \\
& (1-\delta) K_t + \phi \left(\frac{I_t}{K_t} \right) K_t - K_{t+1} \geq 0 
\end{align*}

The TVCs are: 
$$
\lim_{t \rightarrow \infty} \mathbb E_0 \beta^t \frac{u_c (C_{t})}{u_c (C_0)} \frac{B_t}{P_t} = 0, \qquad \lim_{t \rightarrow \infty} \mathbb E_0 \beta^t \frac{u_c (C_{t})}{u_c (C_0)} \frac{K_t}{P_t} = 0
$$

Define the stochastic discount factor for nominal payoffs as $\Lambda_{t,t+1} \equiv \beta \frac{\lambda_{t+1}}{\lambda_t} = \beta \frac{ u_c ( C_{t+1} ) }{u_c (C_{t})} \frac{ P_t }{ P_{t+1} }$, and define $Q_t \equiv \frac{\gamma_t}{\lambda_t P_t}$ as the real shadow value of capital (Tobin's Q). Then, plugging in $\lambda_t$ from [$C_{t}$] into [$N_{t}$], [$B_{t}$], and [$K_{t+1}$], and [$I_{t}$] in terms of $Q_t$ becomes: 
\begin{alignat*}{2}
\text{[Labor supply] :} \quad & \frac{N_{t}^\psi}{C_t^{-\sigma}} &&= \frac{W_t}{P_t} \\
\text{[Euler equation] :} \quad & 1 &&= R_t \mathbb E_t \Lambda_{t,t+1} \\
\text{[$I_{t}$] :} \quad & Q_t  &&= \frac{1}{\phi_{t}'} \\
\text{[$K_{t+1}$] :} \quad & \gamma_t &&= \beta \mathbb E_t \lambda_{t+1} R_{t+1}^K + \beta \mathbb E_t \gamma_{t+1} \left[ 1 - \delta + \phi_{t+1} - \frac{I_{t+1}}{K_{t+1}} \phi_{t+1}' \right] \\
\Longrightarrow\quad & \frac{\gamma_t}{\lambda_t} &&= \beta \mathbb E_t \frac{\lambda_{t+1}}{\lambda_t} R_{t+1}^K + \beta \mathbb E_t \frac{\gamma_{t+1}}{\lambda_t} \left[ 1 - \delta + \phi_{t+1} - \frac{I_{t+1}}{K_{t+1}} \phi_{t+1}' \right] \\
\Longrightarrow\quad & P_t Q_t &&= \mathbb E_t \Lambda_{t,t+1} R_{t+1}^K + \beta \mathbb E_t \frac{Q_{t+1}\lambda_{t+1}P_{t+1}}{\lambda_t} \left[ 1 - \delta + \phi_{t+1} - \frac{I_{t+1}}{K_{t+1}} \phi_{t+1}' \right] \\
& &&= \mathbb E_t \Lambda_{t,t+1} R_{t+1}^K + \mathbb E_t P_{t+1}Q_{t+1} \Lambda_{t,t+1} \left[ 1 - \delta + \phi_{t+1} - \frac{I_{t+1}}{K_{t+1}} \phi_{t+1}' \right] \\
& &&= \mathbb E_t \Lambda_{t,t+1} \left\{ R_{t+1}^K + P_{t+1}Q_{t+1} \left[ 1 - \delta + \phi_{t+1} - \frac{I_{t+1}}{K_{t+1}} \phi_{t+1}' \right] \right\}
\end{alignat*}

Note that the elasticity of investment-capital ratio with respect to Q is: 
\begin{alignat*}{2}
\text{[$I_{t}$]} & \quad\Longrightarrow\quad \ln Q_t  &&= - \ln \phi' \left( \frac{I_t}{K_t} \right) \\
& \quad\Longrightarrow\quad \frac{dQ_t}{Q_t} && = - \frac{ \phi'' \left( \frac{I_t}{K_t} \right) d\left( \frac{I_t}{K_t} \right)}{\phi' \left( \frac{I_t}{K_t} \right)} \\
& \quad\Longrightarrow\quad \frac{d\left( I_t / K_t \right) / \left(I_t / K_t \right)}{dQ_t / Q_t} && = - \frac{\phi' \left( \frac{I_t}{K_t} \right)}{ \phi'' \left( \frac{I_t}{K_t} \right) \left(I_t / K_t \right) } \\
& \quad\Longrightarrow\quad \gamma &&\equiv \left. \frac{d\left( I_t / K_t \right) / \left(I_t / K_t \right)}{dQ_t / Q_t} \right\vert_{I_t = I, K_t = K, Q_t = Q} = - \frac{1}{\phi''(\delta) \delta}
\end{alignat*}

Thus, the optimality conditions of the household are: 
\begin{alignat*}{2}
\text{[Labor supply] :} \quad & \frac{N_{t}^\psi}{C_t^{-\sigma}} &&= \frac{W_t}{P_t} \\
\text{[Euler equation] :} \quad & 1 &&= R_t \mathbb E_t \Lambda_{t,t+1} \\
\text{[$I_{t}$] :} \quad & Q_t  &&= \frac{1}{\phi' \left(\frac{I_t}{K_t}\right)} \\
\text{[$K_{t+1}$] :} \quad & P_t Q_t &&= \mathbb E_t \Lambda_{t,t+1} \left\{ R_{t+1}^K + P_{t+1}Q_{t+1} \left[ 1 - \delta + \phi_{t+1} - \frac{I_{t+1}}{K_{t+1}} \phi_{t+1}' \right] \right\}
\end{alignat*}

#### 2.1. Firms

The firm block of the model are identical to that of Gali et al. (2004). Their derivation are summarized as follows. 

##### 2.2. Final Goods Firm 

A perfectly competitive final goods producer has CRS technology taking $X_t(j)$ as input for each intermediate good $j$. 
$$
Y_t = \left( \int_0^1 X_t(j)^{\frac{\epsilon -1}{\epsilon}} dj \right)^{\frac{\epsilon}{\epsilon - 1}}
$$

Profit maximization given $P_t$ and $P_t(j) \forall j \in [0,1]$ yields optimal demand for each intermediate good $j$: 
$$
X_t(j) = \left( \frac{P_t(j)}{P_t} \right)^{-\varepsilon} Y_t
$$

Zero-profit condition yields: 
$$
P_t = \left( \int_0^1 P_t(j)^{1-\varepsilon} dj \right)^{\frac{1}{1-\varepsilon}}
$$

#### 2.3. Intermediate Goods Firm
The production function for an intermediate goods firm producing good $j$ is: 
$$
Y_t(j) = A_t K_t(j)^\alpha N_t(j)^{1-\alpha}
$$
where $A_t$ is total factor productivity, and $K_t(j)$ is capital rented and $N_t(j)$ is labor hired by firm $j$. 

##### 2.3.1. Cost minimization
The cost minimization problem given wage $W_t$, rental cost of capital $R^k_t$, and technology is given by: 
\begin{align*}
& \min_{N_t(j), K_t(j)} W_t N_t(j) + R_t^k K_t(j) \quad \text{s.t.} \quad Y_t(j) = A_t K_t(j)^\alpha N_t(j)^{1-\alpha} \\
\Longrightarrow & \quad \mathcal{L} \quad = W_t N_t(j) + R_t^k K_t(j) + \lambda_t \left[ Y_t(j) - A_t K_t(j)^\alpha N_t(j)^{1-\alpha} \right]
\end{align*}

The first order conditions are: 
\begin{alignat*}{2}
\text{[$N_{t}$] :} \quad & W_t &&= (1-\alpha) \frac{Y_t(j)}{N_t(j)} \lambda_t \\
\text{[$K_{t}$] :} \quad &  R_t^k &&= \alpha \frac{Y_t(j)}{K_t(j)} \lambda_t 
\end{alignat*}

Combining [$N_{t}$] and [$K_{t}$] yields the optimal capital-labor ratio and the real marginal cost of producing one unit of intermediate good $MC_t$: 
\begin{alignat*}{1}
& \frac{K_t(j)}{N_t(j)}  = \left( \frac{\alpha}{1-\alpha} \right) \left( \frac{W_t}{R^k_t} \right) \\
\Longrightarrow 
& \begin{cases} 
\quad (1) \quad \lambda_t = \frac{1}{\Phi A_t} \left( R_t^k \right)^\alpha W_t^{1-\alpha} \\
\quad\qquad \Longrightarrow MC_t \equiv \frac{\lambda_t}{P_t} = \frac{1}{\Phi A_t} \left( \frac{R_t^k}{P_t} \right)^\alpha \left( \frac{W_t}{P_t} \right)^{1-\alpha} \\
\quad (2) \quad K_t = \left( \frac{\alpha}{1-\alpha} \right) \left( \frac{W_t}{R^k_t} \right) N_t \\
\end{cases} \\
\end{alignat*}

where $\Phi \equiv \alpha^\alpha (1-\alpha)^{1-\alpha} $

##### 2.3.2. Price Setting

Intermediate firms set nominal prices in a staggered fashion ala Calvo (1983). Defining $\theta$ as the fraction of firms keeping prices unchanged, the price-resetting firm seeks to maximize: 
$$
\max_{\{P_t^*\}} E_t \sum_{k=0}^{\infty} \theta^k E_t \left\{ \Lambda_{t,t+k} Y_{t+k}(j) \left( P_t^* - P_{t+k} MC_{t+k} \right) \right\} \quad \text{s.t.} \quad Y_{t+k}(j) = X_{t+k}(j) = \left( \frac{P_t^*}{P_{t+k}} \right)^{-\varepsilon} Y_{t+k}
$$

The first order condition is: 
$$
\sum_{k=0}^{\infty} \theta^k E_t \left\{ \Lambda_{t,t+k} Y_{t+k}(j) \left( P_t^* - \frac{\varepsilon}{\varepsilon - 1} P_{t+k} MC_{t+k} \right) \right\} = 0
$$

The dynamics of aggregate price level is: 
$$
P_t = \left[ \theta P_{t-1}^{1-\varepsilon} + (1 - \theta) (P_t^*)^{1-\varepsilon} \right]^{\frac{1}{1-\varepsilon}}
$$

#### 3. Central Bank 

The central bank sets nominal interest rate $r_t \equiv R_t - 1$ according to the following Taylor rule: 
$$
r_t = r + \phi_\pi \pi_t + \phi_y y_t + \nu_t 
$$

where $\phi_\pi \leq 0, \phi_y \leq 0, r$ is steady state nominal interest rate, and $\nu_t$ is monetary policy shock. 

#### 4. Market Clearing 

\begin{alignat*}{2}
\text{[Labor market] :} \quad & N_t &&= \int_0^t N_t(j) dj \\
\text{[Capital market] :} \quad &  K_t &&= \int_0^t K_t(j) dj \\
\text{[Intermediate goods market] :} \quad &  Y_t(j) &&= X_t(j) \\
\text{[Final goods market] :} \quad &  Y_t &&= C_t + I_t 
\end{alignat*}


#### 5. Log-linearization 

#### 5.1. Household block

Note that in the steady state with $P=1$, 
\begin{alignat}{2}
\text{[Euler equation]} \quad\Longrightarrow\quad & R &&= \frac{1}{\beta} \notag \\
\text{[Law of motion for capital]} \quad\Longrightarrow\quad & \delta K &&= \phi(\frac{I}{K}) K \notag \\
\quad\Longrightarrow\quad & \delta &&= \phi(\frac{I}{K}) \notag \\
\quad\Longrightarrow\quad & \frac{I}{K} &&= \delta \quad \because  \phi(\delta) = \delta \notag \\
\text{[$I_{t}$]} \quad\Longrightarrow\quad & Q &&= \frac{1}{\phi'(\delta)} = 1 \notag \\
\text{[$K_{t+1}$]} \quad\Longrightarrow\quad & PQ &&= \beta \left[ R^K + P (1-\delta + \delta - \delta) \right] \notag \\
\quad\Longrightarrow\quad & \frac{1}{\beta} &&= R^K + (1-\delta) \\ 
\quad\Longrightarrow\quad & R^K &&= \frac{1}{\beta} - (1-\delta) \\
\end{alignat}

Thus, the log-linearized system of equations for the household block becomes: 
\begin{alignat*}{2}
\text{[Labor supply]} \quad\Longrightarrow\quad & \psi \hat{n}_t - \sigma \hat{c}_t &&= \hat{w}_t - \hat{p}_t \\
\text{[Euler equation]} \quad\Longrightarrow\quad & \hat{c}_t &&= \mathbb{E}_t \hat{c}_{t+1} - \frac{1}{\sigma} \left[ \hat{r}_t - \mathbb{E}_t \hat{\pi}_{t+1} \right] \\
\text{[Law of motion for capital]} \quad\Longrightarrow\quad & \hat{k}_{t+1} &&= \delta \hat{i}_t + (1-\delta) \hat{k}_t \\
\text{[$I_{t}$]} \quad\Longrightarrow\quad & \ln Q_t  &&= -\ln \phi' \left( \frac{I_t}{K_t} \right) \\
\quad\Longrightarrow\quad & \frac{dQ_t}{Q} &&= -\frac{\phi''(\delta)}{\phi'(\delta)} \left[ \frac{1}{K} dI_t - \frac{I}{K^2} dK_t  \right] \\ 
\quad\Longrightarrow\quad & \hat{q}_t &&= \frac{1}{\eta} \left[ \hat{i}_t - \hat{k}_t \right] \\
\text{[$K_{t+1}$] / [Euler equation]} \quad\Longrightarrow\quad & \frac{P_t Q_t}{1} &&= \frac{\mathbb E_t \Lambda_{t,t+1} \left\{ R_{t+1}^K + P_{t+1}Q_{t+1} \left[ 1 - \delta + \phi_{t+1} - \frac{I_{t+1}}{K_{t+1}} \phi_{t+1}' \right] \right\} }{R_t \mathbb E_t \Lambda_{t,t+1}} \\ 
\quad\Longrightarrow\quad & \ln P_t - \ln Q_t &&= \ln \mathbb E_t \Lambda_{t,t+1} + \ln \mathbb E_t \left\{ R_{t+1}^K + P_{t+1}Q_{t+1} \left[ 1 - \delta + \phi_{t+1} - \frac{I_{t+1}}{K_{t+1}} \phi_{t+1}' \right] \right\} - \ln R_t - \ln \mathbb E_t \Lambda_{t,t+1} \\
\quad\Longrightarrow\quad & \frac{dP_t}{P} - \frac{dQ_t}{Q} &&= \beta \left\{ d\mathbb{E}_t R_{t+1}^K + \left(d\mathbb{E}_t P_{t+1} + d\mathbb{E}_t Q_{t+1}\right) \left( 1 - \delta \right) - \delta^2 \phi''(\delta) (\mathbb{E}_t \hat{i}_{t+1} - \mathbb{E}_t \hat{k}_{t+1}) \right\} - \hat{r}_t \qquad \text{... (1) used} \\
\quad\Longrightarrow\quad & \hat{p}_t + \hat{q}_t &&= \beta R^k \mathbb{E}_t \hat{r}_{t+1}^K + \beta (1-\delta) \mathbb{E}_t \hat{p}_{t+1} + \beta (1-\delta) \mathbb{E}_t \hat{q}_{t+1} - \beta \delta \frac{1}{\eta} (\eta \mathbb{E}_t \hat{q}_{t+1}) - \hat{r}_t \\
\quad\Longrightarrow\quad & \hat{p}_t + \hat{q}_t &&= \left[ 1-\beta(1-\delta) \right] \mathbb{E}_t \hat{r}_{t+1}^K + \beta (1-\delta) \mathbb{E}_t \hat{p}_{t+1} + \beta (1-\delta) \mathbb{E}_t \hat{q}_{t+1} - \beta \delta \mathbb{E}_t \hat{q}_{t+1} - \hat{r}_t  \qquad \text{... (2) used} \\
\quad\Longleftrightarrow\quad & \hat{q}_t &&= \beta \mathbb{E}_t \hat{q}_{t+1} + \left[ 1- \beta(1-\delta)  \right]  \mathbb{E}_t ( \hat{r}_{t+1}^K - \hat{p}_{t+1})- (\hat{r}_t - \mathbb{E}_t \hat{\pi}_{t+1}) \\
\end{alignat*}

#### 5.2. Firm block

Using the same log-linearization procedure as that of Gali et al. (JMCB 2004) for the price-setting equation, aggregate price dynamics, and aggregate production and log-linearizing the marginal cost of production yield:
\begin{alignat*}{2}
\text{[Phillips curve] :} \quad & \hat{\pi}_t &&= \frac{(1-\theta)(1-\beta\theta)}{\theta} \hat{mc}_t + \beta \mathbb{E}_t \hat{\pi}_{t+1} \\
\text{[Marginal cost] :} \quad & \hat{mc}_t &&= \alpha \hat{r}_t^K + (1-\alpha)\hat{w}_t - \hat{p}_t - a_t \\
\text{[Aggregate production] :} \quad & \hat{y}_t &&= a_t + \alpha \hat{k}_t + (1-\alpha) \hat{n}_t
\end{alignat*}

where t.i.p. stands for terms independent of policy. 

#### 5.3. Central Bank 

The Taylor rule in terms of deviation from steady state is: 
$$
\text{[Monetary policy] :} \quad \hat{r}_t = \phi_\pi \hat{\pi}_t + \phi_y \hat{y}_t + \nu_t
$$

#### 5.4. Market clearing
As in Gali at al. (JMCB 2004), log-linearization of the final goods market clearing condition yields: 
$$
\text{[Final goods market] :} \quad \hat{y}_t = \omega_c \hat{c}_t + (1-\omega_c) \hat{i}_t
$$

where $\omega_c \equiv \frac{C}{Y} = 1 - \frac{\delta \alpha (1-\frac{1}{\epsilon})}{\rho + \delta}$, where $\rho \equiv -\ln \beta$. 

#### 5.5. Shock processes

\begin{align*}
\text{[Technology shock] :} \quad a_t &= \rho_a a_{t-1} + e_{a,t} \\
\text{[Monetary policy shock] :} \quad \nu_t &= \rho_\nu \nu_{t-1} + e_{\nu,t}
\end{align*}

#### 6. The DAG

The 12 equations of the New Keynesian model with endogenous capital accumulation can be organized as four blocks. These 12 equations and the two shock processes will uniquely pin down the equilibrium. 

1. Firm block:
   - $ \hat{\pi}_t = \frac{(1-\theta)(1-\beta\theta)}{\theta} \hat{mc}_t + \beta \mathbb{E}_t \hat{\pi}_{t+1} $
   - $ \hat{mc}_t = \alpha \hat{r}_t^K + (1-\alpha)\hat{w}_t - \hat{p}_t - a_t $
   - $ \hat{k}_t = \hat{w}_t - \hat{r}_t^K + \hat{n}_t $
   - $ \hat{y}_t = a_t + \alpha \hat{k}_t + (1-\alpha) \hat{n}_t $
2. Household block:
   - $ \psi \hat{n}_t - \sigma \hat{c}_t = \hat{w}_t - \hat{p}_t $
   - $ \hat{c}_t = \mathbb{E}_t \hat{c}_{t+1} - \frac{1}{\sigma} \left[ \hat{r}_t - \mathbb{E}_t \hat{\pi}_{t+1} \right] $ 
   - $ \hat{k}_{t+1} = \delta \hat{i}_t + (1-\delta) \hat{k}_t $
   - $ \hat{q}_t = \frac{1}{\eta} \left[ \hat{i}_t - \hat{k}_t \right] $
   - $ \hat{q}_t = \beta \mathbb{E}_t \hat{q}_{t+1} + \left[ 1- \beta(1-\delta)  \right]  \mathbb{E}_t ( \hat{r}_{t+1}^K - \hat{p}_{t+1})- (\hat{r}_t - \mathbb{E}_t \hat{\pi}_{t+1}) $
   - $ \hat{\pi}_{t+1} = \hat{p}_{t+1} - \hat{p}_{t} $
3. Monetary policy
   - $ \hat{r}_t = \phi_\pi \hat{\pi}_t + \phi_y \hat{y}_t + \nu_t $
3. Market clearing block:   
   - $ \hat{y}_t = \omega_c \hat{c}_t + (1-\omega_c) \hat{i}_t $
   

The following rearrangement forms a DAG in two unknowns $\mathbf{U} = \{\mathbf{\hat{k}}, \mathbf{\hat{n}} \}$ and one target, the Euler equation and the dynamics of Tobin's Q. Denote real variables with subscript $R$. 

1. Household investment block: 
    - $ \hat{i}_t = \frac{1}{\delta} \hat{k}_{t+1} - \frac{1-\delta}{\delta} \hat{k}_t $
    - $ \hat{q}_t = \frac{1}{\eta} \left[ \hat{i}_t - \hat{k}_t \right] $
2. Aggregate resource constraint: 
    - $ \hat{y}_t = a_t + \alpha \hat{k}_t + (1-\alpha) \hat{n}_t $
3. Market clearing: 
    - $ \hat{c}_t = \frac{1}{\omega_c} \hat{y}_t - \frac{1-\omega_c}{\omega_c} \hat{i}_t $
4. Household labor supply: 
    - $ \hat{w}_{R,t} = \psi \hat{n}_t - \sigma \hat{c}_t$
5. Firm block: 
    - $ \hat{r}_{R,t}^K = \hat{w}_{R,t}  + \hat{n}_t - \hat{k}_t$
    - $ \hat{mc}_t = \alpha \hat{r}_{R,t}^K + (1-\alpha)\hat{w}_{R,t} - a_t $
    - $ \hat{\pi}_t - \beta \mathbb{E}_t \hat{\pi}_{t+1} = \frac{(1-\theta)(1-\beta\theta)}{\theta} \hat{mc}_t $
6. Monetary policy: 
    - $ \hat{r}_t = \phi_\pi \hat{\pi}_t + \phi_y \hat{y}_t + \nu_t $
7. Euler equation and dynamics of Tobin's Q: 
    - $ 0 = \mathbb{E}_t \hat{c}_{t+1} - \hat{c}_t - \frac{1}{\sigma} \left[ \hat{r}_t - \mathbb{E}_t \hat{\pi}_{t+1} \right] $ 
    - $ 0 = \beta \mathbb{E}_t \hat{q}_{t+1} - \hat{q}_t + \left[ 1- \beta(1-\delta)  \right]  \mathbb{E}_t \hat{r}_{R, t+1}^K - (\hat{r}_t - \mathbb{E}_t \hat{\pi}_{t+1}) $

Note that this DAG can be collpased into an RBC model with capital accumulation, by imposing $\hat{\pi}_t = 0$, $\hat{q}_t = 0$, and combining the Euler equation and Tobin's Q dynamics, which yields the following Euler equation under flexible price equilibrium: 
$$
0 = \left[ 1- \beta(1-\delta)  \right]  \mathbb{E}_t \hat{r}_{t+1}^K - \sigma (\mathbb{E}_t \hat{c}_{t+1} - \hat{c}_t)
$$  

The DAG can also be collapsed into an RBC model without capital accumulation by setting $\alpha = 0$. 

```python
# Import packages
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from SeqSpaceNK import solveSeqSpaceNK
import time 
```

```python
# plot IRFs to aggregate supply shock with persistence rho_a
T = 100 # time horizon needs to be longer for convergence to steady state
rho_a  = 0.9

nu = np.zeros((T, 1))

a = np.zeros((T, 1))
a[0] = 0.01
for t in range(1, T):
    a[t] = rho_a * a[t-1]

# Stack shocks into a single column
shocks = np.vstack((a, nu))

# Loop over models
# for model, store the impulse response functions 11*T 
IRFs = np.zeros((11*T, 3))

# RBC without capital
ss = solveSeqSpaceNK(horizon=T, theta = 0.00001, alpha = 0)
X = ss.solve_seqspace(shocks)
X = X.flatten()
IRFs[:,0] = X

# RBC with capital
ss = solveSeqSpaceNK(horizon=T, theta = 0.00001)
X = ss.solve_seqspace(shocks)
X = X.flatten()
IRFs[:,1] = X

# New Keynesian with capital
ss = solveSeqSpaceNK(horizon=T)
X = ss.solve_seqspace(shocks)
X = X.flatten()
IRFs[:,2] = X

```

```python
# Unpack (k, n, i, q, y, c, w_real, rk_real, mc_real, pi, r) and plot IRFs
k = IRFs[0:T, :]
n = IRFs[T:2*T, :]
i = IRFs[2*T:3*T, :]
q = IRFs[3*T:4*T, :]
y = IRFs[4*T:5*T, :]
c = IRFs[5*T:6*T, :]
w_real = IRFs[6*T:7*T, :]
rk_real = IRFs[7*T:8*T, :]
mc_real = IRFs[8*T:9*T, :]
pi = IRFs[9*T:10*T, :]
r = IRFs[10*T:11*T, :]

# Calculate r_real as r - pi(+1)
pi_plus_1 = np.vstack([pi[1:], np.zeros((1, pi.shape[1]))])
r_real = r - pi_plus_1

# Unpack IRFs
variables = [pi, y, c, k, i, q, n, w_real, rk_real, mc_real, r_real, r]
labels = ['Inflation', 'Output', 'Consumption', 'Capital', 'Investment', 'Tobin\'s q', 'Employment', 'Real wage', 'Real rental rate', 'Marginal cost', 'Real interest rate', 'Nominal interest rate']

# Create a time index for the x-axis
time_index = np.arange(T)

# Plot IRFs
fig, axs = plt.subplots(6, 2, figsize=(15, 20))
for idx, ax in enumerate(axs.flat):
    if labels[idx] in ['Capital', 'Investment', 'Tobin\'s q', 'Real rental rate']:
        ax.plot(time_index, variables[idx][:, 1]*100, label='RBC with capital')
        ax.plot(time_index, variables[idx][:, 2]*100, label='NK with capital')
        ax.legend(['RBC with capital', 'New Keynesian with capital'])
    else:
        ax.plot(time_index, variables[idx][:, 0]*100, label='RBC without capital')
        ax.plot(time_index, variables[idx][:, 1]*100, label='RBC with capital')
        ax.plot(time_index, variables[idx][:, 2]*100, label='NK with capital')
        ax.legend(['RBC without capital', 'RBC with capital', 'New Keynesian with capital'])
    ax.set_title(labels[idx])
    ax.set_ylabel('% from SS deviation')
    ax.set_xlabel('Time')
plt.tight_layout()
plt.show()
```
